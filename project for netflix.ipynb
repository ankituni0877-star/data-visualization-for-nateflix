```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('/home/codespace/data-visualization-for-nateflix/netflix_titles.csv')
# print(df.head())
df = df.dropna(subset=['show_id','type','title','director','cast','country','date_added','release_year','rating','duration','listed_in','description'])
df.drop(columns=['description'])
# print(df.head(5))


# x=df['type']
# y =df['type'].value_counts()
type_counts =df['type'].value_counts()
plt.figure(figsize=(6,4))
plt.bar(type_counts.index ,type_counts.values ,color=['red' , 'orange'])
plt.title('no of movies vs tv shows')
plt.xlabel("Type")
plt.ylabel('count')
plt.tight_layout()




rating_counts = df['rating'].value_counts()
plt.figure(figsize=(8,6))
plt.pie(rating_counts,labels=rating_counts.index,autopct='%1.1f%%',startangle=90)
plt.title('percentage of content')
plt.tight_layout()



movies=df[df['type']== 'Movie'].copy()
movies['duration_int']=movies['duration'].str.replace('min','').astype(int)

plt.figure(figsize=(8,6))
plt.hist(movies['duration_int'] ,bins=30,color='grey',edgecolor='black')
plt.title('movie durations')
plt.xlabel('durations')
plt.ylabel('no of movies')
plt.tight_layout()



release_counts=df['release_year'].value_counts().sort_index()
plt.figure(figsize=(8,6))
plt.scatter(release_counts.index, release_counts.values ,color='red')
plt.title("scatter plot")
plt.xlabel('year')
plt.ylabel('no of movies')
plt.tight_layout()



country_counts=df['country'].value_counts().head(10)
plt.figure(figsize=(6,8))
plt.barh(country_counts.index,country_counts.values, color='blue')
plt.title('top coutries watching shows')
plt.xlabel('no of shows')
plt.ylabel('country name')
plt.tight_layout()



content_by_year= df.groupby(['release_year','type']).size().unstack().fillna(0)

fig,ax = plt.subplots(1,2,figsize=(12,5))

ax[0].plot(content_by_year.index ,content_by_year['Movie'],color='red')
ax[0].set_title('TV SHOWS per year')
ax[0].set_xlabel('year')
ax[0].set_ylabel('no of movies')

ax[1].plot(content_by_year.index ,content_by_year['TV Show'],color='red')
ax[1].set_title('movies_released per year')
ax[1].set_xlabel('year')
ax[1].set_ylabel('no of movies')

fig.suptitle('movies and tv show')
plt.tight_layout()

plt.show()


```
